### Split element and create a new column

```python


# Column to split
column_to_split = "Name"

# Separator for splitting
separator = ","

# New column name
new_column = "First name"

# Loop through each inner dictionary
for key, inner_dict in df_dictionary.items():
    # Split the value of the specified column
    split_values = inner_dict[column_to_split].split(separator)

    # Assign the first split value to the original column
    inner_dict[column_to_split] = split_values[0]

    # Create the new column and assign the second split value
    inner_dict[new_column] = split_values[1]

# Print the updated dictionary
df_dictionary
```

### replace element in a dictionary

```python
# Key, subkey, and new value
key = 0
subkey = "Name"
new_value = "Braundo"

# Directly replace the value of the subkey within the inner dictionary
df_dictionary[key][subkey] = new_value
```
